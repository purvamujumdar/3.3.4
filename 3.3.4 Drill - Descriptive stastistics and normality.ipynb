## The code below shows six sifferent types of distributions and their histograms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
binomial = np.random.binomial(50, 0.9, 1000)
plt.hist(binomial)
plt.axvline(binomial.mean(), color='red', linestyle = 'solid')
plt.axvline(binomial.mean() + binomial.std(), color = 'red', linestyle = 'dashed')
plt.axvline(binomial.mean() - binomial.std(), color = 'red', linestyle = 'dashed')
plt.show()
```

```python
exponential = np.random.exponential(2,1000)
plt.hist(exponential)
plt.axvline(exponential.mean(), color='red', linestyle = 'solid')
plt.axvline(exponential.mean() + exponential.std(), color='red', linestyle = 'dashed')
plt.axvline(exponential.mean() - exponential.std(), color='red', linestyle = 'dashed')
plt.show()
```

```python
poisson = np.random.poisson(8,1000)
plt.hist(poisson)
plt.axvline(poisson.mean(), color='red', linestyle = 'solid')
plt.axvline(poisson.mean() + poisson.std(), color='red', linestyle = 'dashed')
plt.axvline(poisson.mean() - poisson.std(), color='red', linestyle = 'dashed')
plt.show()
```

```python
uniform = np.random.uniform(10, 50, 1000)
plt.hist(uniform)
plt.axvline(uniform.mean(), color='red', linestyle = 'solid')
plt.axvline(uniform.mean() + uniform.std(), color='red', linestyle = 'dashed')
plt.axvline(uniform.mean() - uniform.std(), color='red', linestyle = 'dashed')
plt.show()
```

```python
lognormal = np.random.lognormal(50,25,1000)
plt.hist(lognormal)
plt.axvline(lognormal.mean(), color='red', linestyle = 'solid')
plt.axvline(lognormal.mean() + lognormal.std(), color='red', linestyle = 'dashed')
plt.axvline(lognormal.mean() - lognormal.std(), color='red', linestyle = 'dashed')
plt.show()
```

```python
rayleigh = np.random.rayleigh(8,1000)
plt.hist(rayleigh)
plt.axvline(rayleigh.mean(), color='red', linestyle = 'solid')
plt.axvline(rayleigh.mean() + rayleigh.std(), color='red', linestyle = 'dashed')
plt.axvline(rayleigh.mean() - rayleigh.std(), color='red', linestyle = 'dashed')
plt.show()
```

## __The code below is for the second part of question. x and y are two normally distributed variables. z is the sum of x and y.__

```python
x = np.random.normal(5,0.5,1000)

y = np.random.normal(10,1,1000)

z = x + y
plt.axvline(z.mean(), color='red', linestyle = 'solid')
plt.axvline(z.mean() + z.std(), color='red', linestyle = 'dashed')
plt.axvline(z.mean() - z.std(), color='red', linestyle = 'dashed')
plt.hist(z)

plt.show()
```

The results are going to vary based on the random numbers generated.
